```python
#opening a text file in python
with open("julyTemps.txt") as infile:
    high = []
    low = []
    for line in infile:
        fields = line.split()
        if len(fields)==3 and fields[0]!="Boston" and fields[0]!="Day":
            high.append(int(fields[1]))
            low.append(int(fields[2]))
```

```python
def producePlot(lowTemps, highTemps):
    if len(lowTemps)!=len(highTemps):
        raise Exception("unequal length")
    diffTemps = [highTemps[i]-lowTemps[i] for i in range(len(lowTemps))]
    pylab.plot(range(1, len(lowTemps)+1))
    pylab.title("Day by Day Ranges in Temperature in Boston in July 2012")
    pylab.xlabel("Days")
    pylab.ylabel("Temperature Ranges")
```

```python
producePlot(high, low)
```
